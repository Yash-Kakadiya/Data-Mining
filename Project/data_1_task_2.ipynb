# Step 1: Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("OnlineRetail.csv", encoding='latin1')
df
```

# Step 2: Drop Missing Invoices or Descriptions

```python
df.dropna(subset=["InvoiceNo", "Description"], inplace=True)
df
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
df
```

# Step 4: Keep Only Positive Quantities

```python
df = df[df['Quantity'] > 0]
df
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
df['Description'] = df['Description'].str.lower().replace(' ', '')
df
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
df = df[df['Country'] == 'United Kingdom']
df
```

# Step 7: Remove Duplicates

```python
df.drop_duplicates(inplace=True)
df
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df
```

# Step 9 : Check for Top 10 Selling Products

```python
df['Description'].value_counts().head(10)
```

# Step 10 : Identify Most Active Customers

```python
most_active_customers = df.groupby('CustomerID')['Quantity'].sum().sort_values(ascending=False).head(10)
most_active_customers
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')

df['Month'] = df['InvoiceDate'].dt.month
df['Hour'] = df['InvoiceDate'].dt.hour
```

```python
monthly_sales = df['Month'].value_counts().sort_index()

monthly_sales.plot(kind='bar', title='Transactions per Month', xlabel='Month', ylabel='Number of Transactions')
plt.show()
```

```python
basket = df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().fillna(0)
```

```python
basket = basket.map(lambda x: 1 if x > 0 else 0)
```
